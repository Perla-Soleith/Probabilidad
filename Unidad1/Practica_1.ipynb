```python
from statistics import *
```

```python
# Lista de calificaciones del grupo de probabilidad y estadística

grupoa =[70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print (grupoa)
type(grupoa)
```

```python
#Sacar la media de las calificaciones con una formula

promedio = sum(grupoa) / len (grupoa)
print(promedio)
```

```python
#Utilizar la funcion mean que esta en la libreria statitics para sacar el promedio

promedio2 = mean(grupoa)
print(promedio2)
```

```python
#Determinar el valor de la mediana con la funcion median

mediana = median(grupoa)
print (mediana)
```

```python
#Determinar el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

```python
#En caso de que exista mas de una moda, se puede emplear la funcion multimode para sacar
#multiples modas que puedan existir


moda2 = multimode(grupoa)
print(moda2)
```

```python
#Definimos una ariable llamada tiempos100mts de tipo lista, que contiene los
#tiempos que tardaron los 10 alumnos pera correr los 100 mtros

tiempos100mts = [15.10, 17.21, 16.42, 20.65, 19.73, 15.34, 12.45, 13.36, 17.56, 14.97]

#Despleamos los valores de la lista
print (tiempos100mts)
```

```python
#Usamos la funcion mean para determinar el tiempo promedio de la lista de valores

print ("El tiempo promedio de los corredores es => ", mean(tiempos100mts))
print ("La mediana de los corredores es => ", median(tiempos100mts))
print ("La moda de los corredores es => ", multimode(tiempos100mts))
```

```python
import random
```

```python
fabrica1 =[]
for i in range(100):
  numero_aleatorio = round (random.uniform(50,100), 2)  #El número aleatorio puuede tener hasta dos decimales
  fabrica1.append(numero_aleatorio)
print (fabrica1)
```

```python
fabrica2 =[]
for i in range(100):
  numero_aleatorio = round (random.uniform(70,100), 2)  #El número aleatorio puede tener hasta dos decimales
  fabrica2.append(numero_aleatorio)
print (fabrica2)
```

```python
#libraia de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la Fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Producción diaria de la Fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print ("VALORES DE LA FÁBRICA 1: ")
print ("La media es =", media1)
print ("La mediana es =", mediana1)
print ("La moda es =", moda1)
```

```python
print ("VALORES DE LA FÁBRICA 2: ")
print ("La media es =", mean(fabrica2))
print ("La mediana es =", median(fabrica2))
print ("La moda es =", mode(fabrica2))
```

```python
#Generar 2 subplot de 1 renglon por dos columnas para desplegar los 2 histogramas
#una al lado de la otra

fig, (ax1, ax2) = plt.subplots (nrows=1, ncols=2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
