```python
#creamos una lista vacía que representa la edad de 100 personas

import random
edades =[]
#Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range (100):
  #generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  #nuemore_aleatorio
  numero_aleatorio = random.randint (18, 70)

  #agregamos el numero aleatorio a la lista de valores
  edades.append (numero_aleatorio)

print (edades)
```

```python
import pandas as pd
```

```python
#crear un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame (edades, columns = ['Edad'])
df.head() #comando que muestra las primeras 5 lineas del archivo
```

```python
#calcular latabla de distribución de frecuencias

tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values (by = 'Edad')

#imprimir la tabla de distribución de frecuencias
print ("Tabla de Distribución de Frecuencias: ")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias ['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias ['Frec porcentual'] = tabla_frecuencias ['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias ['Frec Acumulada'] = tabla_frecuencias ['Frecuencia'].cumsum()
tabla_frecuencias
```

```python
#Regla de sturges


#Importamos la libreria math de funciones matematicas

import math
```

```python
#Volvemos a usar la lista de edades
print(edades)
```

```python
#Crear un DataFrame con las edades
df - pd.DataFrame({'Edad': edades})
df
```

```python
#Calcular el número de intervalos utilizando la regla de Sturges
n = len(edades)
k = int (1+ math.log2(n))
print ("Valor de n = ", n, " \nNúmero de intervalos = ", k)
```

```python
#Crear la tabla de frecuencia

tabla_frecuencias = pd.cut ( df ['Edad'], bins = k, include_lowest = True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
#Ordenar la tabla de frecuencia por el intervalo

tabla_frecuencias = tabla_frecuencias.sort_values (by = 'Clase')
tabla_frecuencias
```

```python
#generar columanas siguientes
tabla_frecuencias ['Frec Rel'] = tabla_frecuencias ['Frec'] / n
tabla_frecuencias ['Frec Porc'] = tabla_frecuencias ['Frec'] / n * 100
tabla_frecuencias ['Frec Ac'] = tabla_frecuencias ['Frec'].cumsum()
tabla_frecuencias ['Frec Rel Ac'] = tabla_frecuencias ['Frec Rel'].cumsum()
tabla_frecuencias ['Frec Porc Ac'] = tabla_frecuencias ['Frec Porc'].cumsum()

#mostar la tabla de frecuencias
print (tabla_frecuencias)
```
