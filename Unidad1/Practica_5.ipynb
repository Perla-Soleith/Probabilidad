```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
#Caragar el archivo CSV en ujna variable llamada covid
covid = pd.read_csv("COVID19.csv", encoding = "ISO-8859-1")
```

```python
#Mostar los primeros 5 registros del archivo
covid.head()
```

```python
#Mostrar la información del archivo
#Muestra el numero de reGistros, tipos de datos de la columna, valores nulos, memoria
covid.info()
```

```python
#Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
#Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
#Dibuja el histograma del atributo "EDAD" con bins = 3
#Calcular la media y guardaarla en una variable para mostrarla en el histograma

media = covid.EDAD.mean()

plt.pyplot.hist (covid.EDAD, bins = 11, color = 'pink')
plt.pyplot.title ("Histograma Edad Pruebas COVID")
plt.pyplot.axvline (media, color = 'black', linestyle = 'dashed', linewidth = 2, label = f'Media= {media:.2f}')
plt.pyplot.xlabel ( "Edad" )
plt.pyplot.ylabel ("Frecuencia")
```

```python
#Dibuja el histograma de la columna sexo,
#esto permite comparar cuantos hay de cada grupo
plt.pyplot.hist (covid.SEXO, bins = 3, color = 'green', alpha = 0.5)

#establece las etiquetas x/y y muestra el título
plt.pyplot.title ("Histograma Sexo de Pruebas de Covid")
plt.pyplot.xlabel ("Sexo")
plt.pyplot.ylabel ("Frecuencia")
```

```python
#Dibuja el histograma del atributo "Tipo Paciente, esto permite
#comparar cuantos hay de cada grupo
plt.pyplot.hist (covid.TIPO_PACIENTE, bins = 3, color = 'gray')

#Establece las etiquetas x/y y muestra el título
plt.pyplot.title ("Histograma de tipos de Pacientes en Pruebas COVID")
plt.pyplot.xlabel ("Tipo de Paciente")
plt.pyplot.ylabel ("Frecuencia")
```

```python
#Crear el diagrama de cajas para visualizar la distribución
#de los rangos de edad por sexo

sns.boxplot (x= "SEXO", y = "EDAD", data = covid, color = 'red')
```

```python
#Crear un diagrama de cajas para visualizar la distribución de
#los rangos de edade por tipo de paciente

sns.boxplot (x="TIPO_PACIENTE", y = "EDAD", data = covid, color = 'purple')
```

```python
#Crear un diagrama de cajas para visualizar la distribución de
#los rangos de edad por tipo de paciente

sns.boxplot (x="RESULTADO", y = "EDAD", data = covid, color = 'orange')
```
